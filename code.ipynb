## convert the mnist dataset into images

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'sign_mnist_train.csv'


# Load data from CSV file
data = pd.read_csv(file_path).values



labels = data[:, 0]  # Extract labels
pixels = data[:, 1:]  # Extract pixel values

# Define the dimensions of the images (assuming they are square)
image_size = int(np.sqrt(pixels.shape[1]))

# Create the 'images' folder if it doesn't exist
if not os.path.exists('images'):
    os.makedirs('images')

# Save the images in the 'images' folder
for i in range(len(labels)):
    label = labels[i]
    image = pixels[i, :].reshape((image_size, image_size))

    plt.imshow(image, cmap='gray')
    plt.title(f"Label: {label}")

    # Save the image in the 'images' folder with a unique filename
    image_filename = f"data/images/images/image_{i}_label_{int(label)}.png"
    plt.savefig(image_filename)
    plt.close()

    print(f"Saved: {image_filename}")
```
